```python
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df = pd.read_csv('Case Study Data.csv')
df.head()
```

```python
df=df.dropna()
df.info()
```

```python

df_plus=df[df['Quantity']>=0]
#df_plus=df_plus[df_plus['Invoice Date']>='2018-07-01']
df.shape[0]
df_plus.shape[0]
```

```python
len(pd.unique(df_plus['SKU'])) 
```

```python
basket = df_plus.groupby(['Order Number','Product Category'])['Quantity'].sum().reset_index(name='Quantity')
basket
```

```python
basket.info()
```

```python
basket = basket.pivot(index=['Order Number'], columns='Product Category', values='Quantity').rename_axis(columns=None).fillna(0)
basket
```

```python
def encode_units(x):
    if x<=0:
        return 0
    if x>=1:
        return 1

basket_encode_plus=basket.applymap(encode_units)
basket_encode_plus
```

```python
basket_encode_plus = basket_encode_plus[(basket_encode_plus>0).sum(axis=1)>=2]
basket_encode_plus
```

```python
frequent_items_plus=apriori(basket_encode_plus, min_support=0.03,use_colnames=True) .sort_values('support',ascending=False).reset_index(drop=True)
frequent_items_plus['length']=frequent_items_plus['itemsets'].apply(lambda x: len(x))
frequent_items_plus
```

```python
rules=association_rules(frequent_items_plus, metric="lift", min_threshold=0.5).sort_values(['lift','confidence'],ascending=[False,False]).reset_index(drop=True)
```

```python
rules.to_csv('out.csv')
```
